```python
#2 subplots

figure=plt.figure(figsize=(10,10),facecolor='lightpink')
subplott1=figure.add_subplot(4,2,1,facecolor='lightblue')
subplott2=figure.add_subplot(4,2,2,facecolor='lightblue')
subplott3=figure.add_subplot(4,2,3,facecolor='lightblue')
subplott4=figure.add_subplot(4,2,4,facecolor='lightblue')
a=np.random.randint(1,100,10)
b=np.random.randint(1,100,10)
c=np.random.randint(1,100,10)
d=np.random.randint(1,100,10)
print(a)
print(b)
print(c)
print(d)
subplott1.plot(a)
subplott2.plot(b)
subplott3.plot(c)
subplott4.plot(d)
plt.show()
```

```python
figobj=plt.figure(figsize=(6,6),facecolor='lightblue')##red,green,blue
spobj_1=figobj.add_subplot(1,1,1,facecolor='white')
y=np.random.randint(1,100,10)
x=np.arange(1,20,2)
yt=np.linspace(0,100,10)
xt=np.arange(0,20,2)
y.sort()
print(y)

spobj_1.plot(x,y,linestyle='dashed',drawstyle="steps-mid",marker='$valuemakesiteasy$',mfc='g',mec='blue')
subobj1.set_xticks(xt)
subobj1.set_yticks(yt)
subobj1.set_xlabel("Number")
subobj1.set_ylabel("FReq of rand no")

plt.show()
```

```python
figobj=plt.figure(figsize=(5,5),facecolor='lightblue')##red,green,blue
spobj_1=figobj.add_subplot(1,1,1,facecolor='white')
spobj_1.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
spobj_1.tick_params(axis='y',which='both',bottom=False,top=False,labelbottom=False)
x=np.arange(1,11,1)
y=np.random.randint(1,100,10)
y.sort()
print(y)
y1=y[y<50]
print(y1)
y2=y[y>=50]
print(y2)
y3=np.insert(y2,0,y1[y1.size-1])
print(y3)
y4=np.insert(y3,0,y1[y1.size-2])
print(y4)
print(x)
spobj_1.plot(x[:y1.size],y1,color='black',linestyle='solid',marker="$test$",mfc='blue',mec='green')
spobj_1.plot(x[y1.size-2:],y4,color='red',linestyle='dotted',marker="$red$",mfc='blue',mec='green')

plt.show()
```

```python
figobj=plt.figure(figsize=(5,5),facecolor='lightblue')##red,green,blue
obj=figobj.add_subplot(111,facecolor='white') 
drinks='cola choco rose_berry mrinda'.split()  #delimited is space so the split actions splits based on space, if the space is , or . it has to split in that way
win=[114,14,4,3]
myexplode=[0,0.2,0,0]
color='r y c g'.split()
patches,texts=obj.pie(win,colors=color,shadow=False,startangle=45,explode=myexplode)
obj.set_title("Drinks")
plt.legend(patches,drinks,loc=3,prop={'size':7},fontsize='x=medium')
plt.show()
```

```python
figobj=plt.figure(figsize=(10,10),facecolor='lightblue')
sobj=figobj.add_subplot(111, facecolor='lightpink')
titles='vikky pavi raju bala shanku hema'.split()
me=[0.2,0.4,0,0,0,0]
values=[20,50,60,70,80,100]
color='r,b,g,y'.split(',')
patches,texts=sobj.pie(values,colors=color,startangle=0,explode=me,labels=titles)
sobj.set_title("Family")
plt.legend(titles,prop={'size':10},loc=4)
plt.show()
```

```python
boys_grade=[50,60,60,10,20]
girls_grade=[98,99,80,98,99]
grade_range=np.arange(0,100,10)
print(grade_range)
st=['a','b','c','d','e']
fig=plt.figure(figsize=(6,4),facecolor='lightblue')
ax=fig.add_axes([1,1,1,1,])
ax.axis('equal')
ax.pie(boys_grade,labels=st,autopct='%1.2f%%')
plt.show()
```

```python
#scatter graph

boys_grade=[50,60,60,10,20]
girls_grade=[98,99,80,98,99]
grade_range=np.linspace(0,100,5)
print(grade_range)
st=['a','b','c','d','e']
fig=plt.figure(figsize=(6,4),facecolor='lightblue')
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.scatter(grade_range,boys_grade,color='b',marker="o")
ax.scatter(grade_range,girls_grade,color='g',marker="<")
plt.show()
```
